## L thuyt
Thut ton di truyn (Genetic Algorithm - GA) l mt phng php tm kim da trn c ch tin ha t nhin, ly cm hng t thuyt tin ha ca Darwin. GA c s dng  tm li gii gn ti u cho cc bi ton ti u ha phc tp.

### Cc bc chnh ca thut ton:
1. **Khi to qun th**: To ngu nhin mt tp hp cc c th (gii php tim nng).
2. **nh gi  thch nghi (Fitness Evaluation)**: Tnh gi tr hm mc tiu cho mi c th.
3. **La chn (Selection)**: Chn cc c th tt nht  lai ghp (thng s dng cc phng php nh Tournament Selection hoc Roulette Wheel).
4. **Lai ghp (Crossover)**: Kt hp hai c th  to ra c th mi.
5. **t bin (Mutation)**: Thay i ngu nhin mt s gen  duy tr tnh a dng.
6. **Lp li**: Quay li bc 2 cho n khi t iu kin dng (s th h hoc gi tr fitness mong mun).



### ng dng
- Ti u ha hm s.
- Bi ton lp lch.
- Thit k k thut, machine learning, v.v.

## Cc v d minh ha
Chng ta s thc hin hai v d:

1. Tm gi tr ti a ca hm s $f(x) = -x^2 + 10x + 50$ trong khong $x \in [0, 10]$.
2. Tm gi tr ti thiu ca hm s $g(x, y) = x^2 + y^2$ trong khong $x, y \in [-5, 5]$.

Ngoi ra, chng ta s trc quan ha s thay i ca gi tr fitness qua cc th h.


## V d 1: Ti u ha hm mt bin

### Bi ton  
Tm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
t gi tr ln nht trong khong  
$$
x \in [0, 10].
$$

# M gi
```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection:
        - Chn ngu nhin tournament_size c th trong qun th
        - Chn c th c fitness cao nht lm cha m 1
    Chn cha m 2 bng Tournament Selection tng t
    
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi c gi tr trung bnh ca cha m
        Ngc li:
            Con mi l cha m 1
            
    Thc hin t bin cho con mi vi xc sut mutation_rate:
        Nu t bin:
            Gn con mi gi tr ngu nhin trong [min_val, max_val]
            
    Thm con mi vo qun th mi
    
Cp nht qun th hin ti = qun th mi
In ra c th tt nht v gi tr fitness tng ng ca th h


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hm mc tiu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khi to qun th
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# La chn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# t bin
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thut ton di truyn
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr f(x) tt nht')
    plt.title('S thay i ca f(x) qua cc th h')
    plt.show()
    
    return best_individual, best_fitness

# Chy thut ton
best_x, best_f = genetic_algorithm_example1()
print(f'\nKt qu cui cng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

## V d 2: Ti u ha hm hai bin

### Bi ton  
Tm $(x, y)$ sao cho  
$$
g(x, y) = x^2 + y^2
$$  
t gi tr nh nht trong khong

$$


### M gi 

```plaintext
To qun th mi rng
Lp cho n khi qun th mi  N c th:
    Chn cha m 1 bng Tournament Selection trn qun th hin ti
    Chn cha m 2 bng Tournament Selection trn qun th hin ti
    Thc hin lai ghp (crossover) gia cha m 1 v cha m 2 vi xc sut crossover_rate:
        Nu crossover thnh cng:
            To con mi bng trung bnh gi tr (x, y) ca cha m
        Ngc li:
            Con mi l cha m 1
    Thc hin t bin cho con mi vi xc sut mutation_rate tng bin:
        Nu t bin:
            Gn gi tr ngu nhin trong [min_val, max_val] cho bin tng ng
    Thm con mi vo qun th mi

Cp nht qun th hin ti = qun th mi
In kt qu tt nht th h

````python
# Hm mc tiu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # S dng du m v GA tm max

# Khi to qun th
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# La chn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# t bin
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thut ton di truyn
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyn v gi tr thc
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Th h {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr g(x, y) tt nht')
    plt.title('S thay i ca g(x, y) qua cc th h')
    plt.show()
    
    return best_individual, -best_fitness

# Chy thut ton
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKt qu cui cng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bi tp 1
# Thay i hm mc tiu**: Th p dng thut ton cho hm $h(x) = sin(x) + cos(x)$ trong khong $x \in [0, 2\pi]$. So snh kt qu vi nghim chnh xc.

## Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

## Mc tiu
- Ti u hm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong min:
  $$
  x \in [0, 2\pi]
  $$
- Tm gi tr **x** sao cho **h(x)** t gi tr ln nht.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  tng thut ton di truyn (GA)

1. **M ha**: Biu din gi tr \( x \) bng chui nh phn di 16 bit.  
2. **Gii m**: Chuyn t chui nh phn sang gi tr thc trong khong \([0, 2\pi]\).  
3. **Fitness**: S dng chnh gi tr \( h(x) \) lm  thch nghi.  
4. **Chn lc**: Roulette Wheel Selection.  
5. **Lai ghp**: Lai 1 im vi xc sut cao.  
6. **t bin**: Lt bit vi xc sut thp.

---


````python
import numpy as np
import matplotlib.pyplot as plt
import random

def fitness_function_1d(x):
    return np.sin(x) + np.cos(x)

def initialize_population_1d(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

def select_parents_1d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

def crossover_1d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

def mutate_1d(individual, mutation_rate=0.1, min_val=0, max_val=2*np.pi):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

def genetic_algorithm_hx(pop_size=50, generations=100, min_val=0, max_val=2*np.pi):
    population = initialize_population_1d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_1d(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            p1 = select_parents_1d(population, fitness_values)
            p2 = select_parents_1d(population, fitness_values)
            child = crossover_1d(p1, p2)
            child = mutate_1d(child, min_val=min_val, max_val=max_val)
            new_population.append(child)
        
        population = new_population
        best_x = population[np.argmax([fitness_function_1d(x) for x in population])]
        print(f'Th h {generation + 1}: x = {best_x:.4f}, h(x) = {fitness_function_1d(best_x):.4f}')
    
    best_x = population[np.argmax([fitness_function_1d(x) for x in population])]
    best_h = fitness_function_1d(best_x)
  
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('Tin ha h(x) qua cc th h')
    plt.grid(True)
    plt.show()
    
    return best_x, best_h

best_x, best_h = genetic_algorithm_hx()
print(f'\nKt qu cui cng: x = {best_x:.4f}, h(x) = {best_h:.4f}')
print(f'Gi tr chnh xc: max h(x) = 2  {np.sqrt(2):.4f} ti x = /4  {np.pi/4:.4f}')
````

# Bi tp 2: Ti u hm $h(x) = \sin(x) + \cos(x)$ bng Thut Ton Di Truyn (Genetic Algorithm)

---

## Mc tiu

Ti u hm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khong:
$$
x \in [0, 2\pi]
$$

> Tm gi tr $x$ sao cho $h(x)$ t **gi tr ln nht**.

---

## Phn tch nghim chnh xc

Ta c:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##   tng thut ton di truyn (GA)

| Thnh phn        | Chi tit                                                                 |
|------------------|-------------------------------------------------------------------------|
| **M ha**        | Dng chui nh phn 16 bit  biu din gi tr $x \in [0, 2\pi]$         |
| **Gii m**       | nh x nh phn sang gi tr thc: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chnh l $h(x) = \sin(x) + \cos(x)$                                     |
| **Chn lc**      | Roulette Wheel Selection                                                |
| **Lai ghp**      | 1 im ct vi xc sut 0.8                                              |
| **t bin**      | Xc sut 5% (0.05)                                                       |

---

````python
import numpy as np
import matplotlib.pyplot as plt
import random

# Hm mc tiu
def fitness_function_bin(individual, num_bits=16):
    decimal = int("".join(map(str, individual)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    return np.sin(x) + np.cos(x)

# Khi to qun th
def initialize_population_bin(pop_size, num_bits=16):
    return [np.random.randint(0, 2, num_bits).tolist() for _ in range(pop_size)]

# La chn
def select_parents_bin(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghp
def crossover_bin(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        return parent1[:point] + parent2[point:]
    return parent1[:]

# t bin
def mutate_bin(individual, mutation_rate=0.05):
    return [bit ^ 1 if random.random() < mutation_rate else bit for bit in individual]

# Thut ton di truyn
def genetic_algorithm_example_bin(pop_size=50, generations=100, num_bits=16):
    population = initialize_population_bin(pop_size, num_bits)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_bin(ind, num_bits) for ind in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_bin(population, fitness_values)
            parent2 = select_parents_bin(population, fitness_values)
            child = crossover_bin(parent1, parent2)
            child = mutate_bin(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_bin(ind) for ind in population])]
        decimal = int("".join(map(str, best_individual)), 2)
        x = decimal * (2 * np.pi) / (2**num_bits - 1)
        print(f'Th h {generation + 1}: x = {x:.4f}, h(x) = {fitness_function_bin(best_individual):.4f}')
    
    best_individual = population[np.argmax([fitness_function_bin(ind) for ind in population])]
    decimal = int("".join(map(str, best_individual)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    h_val = fitness_function_bin(best_individual)
    
    # Trc quan ha
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Th h')
    plt.ylabel('Gi tr h(x) tt nht')
    plt.title('S thay i ca h(x) qua cc th h')
    plt.grid(True)
    plt.show()
    
    return x, h_val

# Chy thut ton
best_x, best_h = genetic_algorithm_example_bin()
print(f'\nKt qu cui cng: x = {best_x:.4f}, h(x) = {best_h:.4f}')
print(f'Gi tr chnh xc: max h(x) = 2  {np.sqrt(2):.4f} ti x = /4  {np.pi/4:.4f}')
````

# Bi Tp 3

# Thay i tham s thut ton di truyn v phn tch tc  hi t

**Mc tiu:** Th thay i cc tham s chnh ca thut ton di truyn gm:

- Kch thc qun th: `pop_size`
- S th h: `generations`
- T l t bin: `mutation_rate`
- T l lai ghp: `crossover_rate`

v phn tch nh hng ca tng tham s n tc  hi t ca thut ton.

---

## Cc tham s v vai tr

| Tham s         |  ngha                                  | nh hng chnh                                                                                         |
|-----------------|-----------------------------------------|--------------------------------------------------------------------------------------------------------|
| `pop_size`      | S c th trong qun th                 | Qun th ln gip khm ph khng gian tm kim tt hn nhng tn nhiu thi gian tnh ton. Qun th nh hi t nhanh nhng d b kt cc tr cc b. |
| `generations`   | S vng lp ca thut ton               | S th h cng nhiu th thut ton c nhiu c hi ci thin nghim, tuy nhin thi gian chy cng tng. |
| `mutation_rate` | Xc sut t bin mi bit                | t bin cao gip tng a dng gene, trnh b kt trong cc tr cc b, nhng nu qu cao lm qu trnh tm kim nhiu lon. |
| `crossover_rate`| Xc sut lai ghp gia hai c th       | Lai ghp gip phi hp gen tt, tng kh nng tm nghim tt. Nu qu thp, qun th km a dng; qu cao c th ph hy cu trc tt ca c th. |

---

## Phn tch nh hng

- **`pop_size`**: Kch thc qun th ln gip tm kim ton cc tt hn nhng chm hn, kch thc nh lm thut ton hi t nhanh nhng d mc kt.
- **`generations`**: S th h ln tng kh nng ci tin kt qu nhng lm tng thi gian tnh ton.
- **`mutation_rate`**: T l t bin va phi gip duy tr a dng di truyn, trnh kt cc b; qu cao hoc qu thp u lm gim hiu qu.
- **`crossover_rate`**: T l lai ghp hp l gip khai thc tt thng tin di truyn, tng hiu qu hi t.

---




````python
import numpy as np
import matplotlib.pyplot as plt
import random
import time

def fitness_function_bin(individual, num_bits=16):
    decimal = int("".join(map(str, individual)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    return np.sin(x) + np.cos(x)

def initialize_population_bin(pop_size, num_bits=16):
    return [np.random.randint(0, 2, num_bits).tolist() for _ in range(pop_size)]

def select_parents_bin(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

def crossover_bin(parent1, parent2, crossover_rate):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        return parent1[:point] + parent2[point:]
    return parent1[:]

def mutate_bin(individual, mutation_rate):
    return [bit ^ 1 if random.random() < mutation_rate else bit for bit in individual]

def genetic_algorithm_tunable(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, num_bits=16):
    population = initialize_population_bin(pop_size, num_bits)
    best_fitness_history = []
    start_time = time.time()

    for generation in range(generations):
        fitness_values = [fitness_function_bin(ind, num_bits) for ind in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_bin(population, fitness_values)
            parent2 = select_parents_bin(population, fitness_values)
            child = crossover_bin(parent1, parent2, crossover_rate)
            child = mutate_bin(child, mutation_rate)
            new_population.append(child)

        population = new_population

    best_individual = population[np.argmax([fitness_function_bin(ind, num_bits) for ind in population])]
    decimal = int("".join(map(str, best_individual)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    h_val = fitness_function_bin(best_individual, num_bits)
    elapsed_time = time.time() - start_time

    return x, h_val, best_fitness_history, elapsed_time

configs = [
    {"pop_size": 20, "generations": 50, "mutation_rate": 0.01, "crossover_rate": 0.6},
    {"pop_size": 50, "generations": 100, "mutation_rate": 0.05, "crossover_rate": 0.8},
    {"pop_size": 100, "generations": 100, "mutation_rate": 0.1, "crossover_rate": 0.9},
    {"pop_size": 50, "generations": 200, "mutation_rate": 0.01, "crossover_rate": 0.9},
]

plt.figure(figsize=(10, 6))
for i, cfg in enumerate(configs):
    x_best, h_best, history, elapsed = genetic_algorithm_tunable(**cfg)
    label = f"pop={cfg['pop_size']}, gen={cfg['generations']}, mut={cfg['mutation_rate']}, cross={cfg['crossover_rate']}"
    print(f"Cu hnh {i+1}: {label}")
    print(f"   Kt qu: x = {x_best:.4f}, h(x) = {h_best:.4f}, thi gian = {elapsed:.2f}s\n")
    plt.plot(history, label=label)

plt.xlabel('Th h')
plt.ylabel('Gi tr h(x) tt nht')
plt.title('So snh tc  hi t vi cc tham s GA khc nhau')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Bi tp 4
# Ci tin thut ton  Thm phng php la chn khc v so snh hiu qu

---

## Mc tiu

- Thm phng php la chn mi vo thut ton di truyn, v d nh **Roulette Wheel Selection** (La chn bnh xe con quay).
- So snh hiu qu ca phng php la chn mi vi phng php c (v d: Chn ngu nhin, Tournament Selection, v.v.).
- Phn tch nh hng ca phng php la chn n tc  hi t v cht lng nghim.

---

##  Cc phng php la chn ph bin

| Phng php                | M t                                                                                   | u im                              | Nhc im                          |
|----------------------------|-----------------------------------------------------------------------------------------|------------------------------------|-----------------------------------|
| Chn ngu nhin (Random)   | Chn c th ngu nhin trong qun th                                                   | n gin, nhanh                    | Khng u tin c th tt hn      |
| Roulette Wheel Selection    | Chn theo xc sut t l thun vi  thch nghi (fitness)                              | D dng, u tin c th tt        | D b nh hng bi c th qu mnh|

##  Kt qu v phn tch

- Phng php **Roulette Wheel** thng gip tng tc  hi t nh u tin c th tt.
- Tuy nhin c th lm gim a dng gen nu c th u t chim u th qu ln, d ri vo cc tr cc b.
- Phng php **Tournament** kim sot tt s a dng nhng c th chm hn trong mt s trng hp.
- Phng php **Random** n gin nhng t hiu qu trong ti u nng cao.

````python
import numpy as np
import matplotlib.pyplot as plt
import random
import time

def fitness_function(individual, num_bits=16):
    decimal = int("".join(map(str, individual)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    return np.sin(x) + np.cos(x)

def initialize_population(pop_size, num_bits=16):
    return [np.random.randint(0, 2, num_bits).tolist() for _ in range(pop_size)]

def select_random(population, fitness_values):
    return random.choice(population)

def select_tournament(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

def select_roulette(population, fitness_values):
    min_fit = min(fitness_values)
    if min_fit < 0:
        fitness_values = [f - min_fit + 1e-6 for f in fitness_values]

    total_fit = sum(fitness_values)
    if total_fit <= 0 or any(np.isnan(f) or not np.isfinite(f) for f in fitness_values):
        return random.choice(population)

    probs = [f / total_fit for f in fitness_values]
    return population[np.random.choice(len(population), p=probs)]

def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        return parent1[:point] + parent2[point:]
    return parent1[:]

def mutate(individual, mutation_rate=0.05):
    return [bit ^ 1 if random.random() < mutation_rate else bit for bit in individual]

def genetic_algorithm_selection(method='tournament', pop_size=50, generations=100,
                                mutation_rate=0.05, crossover_rate=0.8, num_bits=16):
    
    if method == 'random':
        selector = select_random
    elif method == 'roulette':
        selector = select_roulette
    else:
        selector = select_tournament

    population = initialize_population(pop_size, num_bits)
    best_fitness_history = []
    start_time = time.time()

    for _ in range(generations):
        fitness_values = [fitness_function(ind, num_bits) for ind in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)

        new_population = []
        for _ in range(pop_size):
            p1 = selector(population, fitness_values)
            p2 = selector(population, fitness_values)
            child = crossover(p1, p2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)

        population = new_population

    best_ind = population[np.argmax([fitness_function(ind, num_bits) for ind in population])]
    decimal = int("".join(map(str, best_ind)), 2)
    x = decimal * (2 * np.pi) / (2**num_bits - 1)
    h_val = fitness_function(best_ind, num_bits)
    elapsed = time.time() - start_time

    return x, h_val, best_fitness_history, elapsed

methods = ['random', 'tournament', 'roulette']
plt.figure(figsize=(10, 6))

for method in methods:
    x_best, h_best, history, elapsed = genetic_algorithm_selection(method=method)
    print(f"Phng php {method.upper()}:")
    print(f"   x = {x_best:.4f}, h(x) = {h_best:.4f}, thi gian = {elapsed:.2f}s\n")
    plt.plot(history, label=f"{method.capitalize()}")

plt.xlabel("Th h")
plt.ylabel("Gi tr h(x) tt nht")
plt.title("So snh tc  hi t gia cc phng php chn lc")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````

# Trc quan ha qun th  V biu  phn tn (scatter plot) th hin s phn b ca qun th qua cc th h

##  Mc tiu

- Trc quan ha s phn b ca qun th c th trong qu trnh tin ha ca thut ton di truyn.
- Dng biu  phn tn (scatter plot)  th hin v tr cc c th trn khng gian bin s qua tng th h.
- Gip quan st r tin trnh hi t, a dng di truyn, v s thay i qun th theo thi gian.

---

##  tng thc hin

- Vi bi ton ti u hm $k(x, y, z) = x^2 + y^2 + z^2$, khng gian tm kim l 3 chiu.
- Mi c th l mt im trong khng gian $(x, y, z)$.
-  mi th h, ly v tr gii m ca ton b c th, v scatter plot.
- Nu khng th trc quan 3D trc tip, c th v scatter 2D tng cp bin: $(x, y)$, $(y, z)$, $(x, z)$ qua cc th h.

---

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random

def fitness_function_3d(x, y, z):
    return -(x**2 + y**2 + z**2)

def initialize_population_3d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val),
             random.uniform(min_val, max_val),
             random.uniform(min_val, max_val)) for _ in range(pop_size)]

def select_parents_3d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

def crossover_3d(p1, p2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return tuple((a + b) / 2 for a, b in zip(p1, p2))
    return p1

def mutate_3d(individual, mutation_rate=0.05, min_val=-10, max_val=10):
    x, y, z = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        z = random.uniform(min_val, max_val)
    return (x, y, z)

def genetic_algorithm_3d(pop_size=50, generations=50, min_val=-10, max_val=10,
                         mutation_rate=0.05, crossover_rate=0.8):

    population = initialize_population_3d(pop_size, min_val, max_val)
    all_generations = []

    for _ in range(generations):
        fitness_values = [fitness_function_3d(x, y, z) for (x, y, z) in population]
        all_generations.append(population.copy())

        new_population = []
        for _ in range(pop_size):
            p1 = select_parents_3d(population, fitness_values)
            p2 = select_parents_3d(population, fitness_values)
            child = crossover_3d(p1, p2, crossover_rate)
            child = mutate_3d(child, mutation_rate, min_val, max_val)
            new_population.append(child)

        population = new_population

    return all_generations

def visualize_3d_generations(generations_to_plot, all_generations):
    for gen in generations_to_plot:
        fig = plt.figure(figsize=(6, 5))
        ax = fig.add_subplot(111, projection='3d')
        generation = all_generations[gen]
        xs = [ind[0] for ind in generation]
        ys = [ind[1] for ind in generation]
        zs = [ind[2] for ind in generation]

        ax.scatter(xs, ys, zs, c='blue', alpha=0.6)
        ax.set_title(f'Phn b qun th  Th h {gen + 1}')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')
        ax.grid(True)
        plt.tight_layout()
        plt.show()

all_generations = genetic_algorithm_3d()

generations_to_plot = [0, len(all_generations)//2, len(all_generations)-1]
visualize_3d_generations(generations_to_plot, all_generations)
````
